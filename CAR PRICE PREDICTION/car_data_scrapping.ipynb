## scrapping details of car

```python
brand=[]
model=[]
year=[]
hist=[]
km=[]
fuel=[]
trans=[]
ownership=[]
loc=[]
price=[]
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
for i in cars[0:2000]:
    driver.get(i)
    time.sleep(2)
    
    brandtag=driver.find_elements(By.XPATH,'//h2[@class="_2geSF"]')
    for i in brandtag:
        brand.append(i.text.split()[1])
        year.append(i.text.split()[0])
        model.append(i.text.split()[2])

# scrapping of general details of the car

    histtag=driver.find_elements(By.XPATH,'//strong[@class="_1-PH2"]')
    for i in histtag[0:1]:
        hist.append(i.text.split()[0])
    for i in histtag[1:2]:
        ownership.append(i.text.split()[0])    

    for i in histtag[2:3]:
        km.append(i.text.split()[0])
    for i in histtag[3:4]:  
        fuel.append(i.text.split()[0])
    for i in histtag[5:6]:
        trans.append(i.text.split()[0])
    
    # scrapping of location
    
    loctag=driver.find_elements(By.XPATH,'//p[@class="_1V4-O"]')
    for i in loctag:
        loc.append(i.text.split())
    
    try:
        pricetag=driver.find_elements(By.XPATH,'//strong[@class="_2yYvS Wqfj6"]')
        for i in pricetag:
            price.append(i.text.splitlines()[1])
    except NoSuchElementException: 
        price.append(np.nan)
```
