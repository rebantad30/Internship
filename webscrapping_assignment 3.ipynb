```python
prodlist=driver.find_elements(By.CLASS_NAME,'_1fQZEK')
```

```python
prod=[]
for i in prodlist:
    prod.append(i.get_attribute('href'))
```

```python
bname=[]
sname=[]
color=[]
ram=[]
rom=[]
primarycam=[]
secondarycam=[]
disp=[]
bat=[]
price=[]
produrl=[]
```

```python
#time.sleep(5)
#for scrapping price 
pricetag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"][1]')
for i in pricetag:
    price.append(i.text)
    
    #scrapping of brand name
bnametag=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"][1]')
for i in bnametag:
    bname.append(i.text.split()[0])
    sname.append(i.text.split()[1])
    color.append(i.text.split()[-4:-2])


ramromtag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][1]')
# this element will scrape details of RAM and ROM
for i in ramromtag:
    ram.append(i.text.split()[0:2])
    rom.append(i.text.split()[4:-1])
# this element will contain the display size of all the products on 1st page
dsptag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][2]')
for i in dsptag:
    disp.append(i.text)  
# this element will contain the information about primary and secondary camera specifications of all the products on 1st page
cameratag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][3]')
for i in cameratag:
    primarycam.append(i.text.split()[0:3])
    secondarycam.append(i.text.split()[6:-1])
# this element will contain the battery capacity information of all the products on 1st page
battag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][4]')
# it has been done to scrape the details of the 4th element from elements with same class name
for i in battag:
    bat.append(i.text.split())
    
urltag=driver.find_elements(By.CLASS_NAME,'_1fQZEK')
for i in urltag:
    # href tag contains the url
    produrl.append(i.get_attribute('href'))
```

```python
df=pd.DataFrame({'Brand Name':bname,'Smartphone model':sname,'Color':color,'RAM':ram,'ROM':rom,'Primary camera':primarycam,'Secondary Camera':secondarycam,'Display':disp,'Price':price,'Bettery capacity':bat,'Product URl':produrl})
```

# Question 10: Write a python program to scrape a data for all available Hostels 

“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews,
overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-09-14&to=2022-09-17&guests=2&page=1')
```

```python
name=[]
price=[]
dist=[]
review=[]
descr=[]
total=[]
nametag=driver.find_elements(By.XPATH,'//div[@class="title-row"]//a')
for i in nametag:
    total.append(i.text.split())
name=total[0::2]

#pricetag=driver.find_elements(By.XPATH,'//div[@class="price title-5"]')
#for i in pricetag:
 #   price.append(i.text)
    
reviewtag=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
for i in reviewtag:
    review.append(i.text.split()[0:1])
    

distag=driver.find_elements(By.XPATH,'//div[@class="subtitle body-3"]')
for i in distag:
    dist.append(i.text.split()[2:6])
# slicing the first 3 duplicate elements
dist=dist[3:33]    

destag=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
for i in destag:
    descr.append(i.text.split())
```

```python
df=pd.DataFrame({'Name':name,'distance':dist,'reviews':review,'Description':descr})
```
